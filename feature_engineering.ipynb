```python
# imports

import numpy as np
import pandas as pd
```

```python
filename = '/content/bikeshare_train - bikeshare_train.csv'
df = pd.read_csv(filename)
df.head()
```

```python
df.drop(columns=['casual', 'registered'], inplace = True)
df.head()
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
df.head()
```

```python
df['Name of the Month'] = df['datetime'].dt.month
df['Name of the Day of the Week'] = df['datetime'].dt.weekday
df['hour of the day'] = df['datetime'].dt.hour
```

```python
df.head()
```

```python
df.info()
```

```python
# converting data types to objects
df = df.astype({"Name of the Month" : object, 'Name of the Day of the Week' : object,
                'hour of the day' : object})
```

```python
df.info()
```

```python
df.drop(columns = ['datetime', 'season'], inplace = True)
```

```python
df['temp'] = df['temp'].apply(lambda x: 9/5 * x + 32)
```

```python
df['atemp'] = df['atemp'].apply(lambda x: 9/5 * x + 32)
```

```python
df.head()
```

```python
df['temp_Variance'] = df['temp'] - df['atemp']
```

```python
# dropping atemp

df = df.drop(['atemp'], axis=1)
```
